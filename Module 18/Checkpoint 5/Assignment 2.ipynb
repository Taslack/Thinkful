```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import statsmodels.api as sm
```

```python
postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'houseprices'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
df = pd.read_sql_query('select * from houseprices',con=engine)

# no need for an open connection, as we're only doing a single query
engine.dispose()
```

```python
# Setting the target varible as Y
Y = df['saleprice']
# Features set as X
X = df[['overallqual', 'grlivarea', 'garagecars', 'garagearea', 'totalbsmtsf', 'firstflrsf', 'fullbath', 'totrmsabvgrd', 'yearbuilt', 'yearremodadd']]

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

- F-test is 495.4 and its p-value is 0.
- R-squared is 0.774 and Adj. R-squared is 0.772.
- AIC is 3.494e+04 and BIC is 3.500e+04.

The F-test p-value is at 0, showing that the model is indistinguishable from the reduced model.  So the model is useful and statistically significant.  Altho R-squared shows that 77.4% of the data is matched to the model, this can possible be improved upon as 22.6% of the variance in the target is not explained.

```python
df['totalsf'] = df['totalbsmtsf'] + df['firstflrsf'] + df['secondflrsf'] +  df['lowqualfinsf'] + df['wooddecksf'] + df['openporchsf']
```

```python
# Setting the target varible as Y
Y = df['saleprice']
# Features set as X
X = df[['overallqual', 'grlivarea', 'garagecars', 'garagearea', 'totalsf']]

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

```python
df['int_*_sf'] = df['totalsf'] * df['overallqual']
```

```python
# Setting the target varible as Y
Y = df['saleprice']
# Features set as X
X = df[['overallqual', 'grlivarea', 'garagecars', 'garagearea', 'totalsf', 'int_over_sf']]

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

```python
# Y is the target variable
Y = np.log1p(df['saleprice'])
# X is the feature set
X = df[['overallqual', 'grlivarea', 'garagecars', 'garagearea', 'totalsf', 'int_*_sf']]

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

The last model is the best, R-squared is 0.810, F-test p-value is 0, AIC is -948.7, and BIC is -911.7.
