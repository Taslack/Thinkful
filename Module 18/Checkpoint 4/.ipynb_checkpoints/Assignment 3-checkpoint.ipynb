```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import statsmodels.api as sm
```

```python
postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'houseprices'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
df = pd.read_sql_query('select * from houseprices',con=engine)

# no need for an open connection, as we're only doing a single query
engine.dispose()
```

```python
# Setting the target varible as Y
Y = df['saleprice']
# Features set as X
X = df[['overallqual', 'grlivarea', 'garagecars', 'garagearea', 'totalbsmtsf', 'firstflrsf', 'fullbath', 'totrmsabvgrd', 'yearbuilt', 'yearremodadd']]

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

- The coefficients that are statistically significant are overallqual, grlivarea, garagecars, totalbsmtsf, firstflrsf, fullbath, yearbuilt, yearremodadd.
- Overallqual is the highest value that is significat, per point increase there is 19600 increase in saleprice.
- Fullbath actully decreases the saleprice, per point decrease by 6790.87.

### Excluding the insignificant features.

```python
# Setting the target varible as Y
Y = df['saleprice']
# Features set as X
X = df[['overallqual', 'grlivarea', 'garagecars', 'totalbsmtsf', 'firstflrsf', 'fullbath', 'yearbuilt', 'yearremodadd']]

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```
