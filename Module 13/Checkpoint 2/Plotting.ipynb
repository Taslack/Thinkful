```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Loading one of the example datasets from seaborn.
exercise = sns.load_dataset("exercise")
print(exercise.head(n=10))
```

```python
lowfatrest1min = exercise.loc[(exercise['diet'] == 'low fat') & (exercise['kind'] == 'rest') & (exercise['time'] == '1 min'), 'pulse']
lowfatrest15min = exercise.loc[(exercise['diet'] == 'low fat') & (exercise['kind'] == 'rest') & (exercise['time'] == '15 min'), 'pulse']
lowfatrest30min = exercise.loc[(exercise['diet'] == 'low fat') & (exercise['kind'] == 'rest') & (exercise['time'] == '30 min'), 'pulse']
print(lowfatrest1min, lowfatrest15min, lowfatrest30min)
```

```python
sns.relplot(x='time', y='pulse', data=exercise)
```

```python
sns.relplot(x='time', y='pulse', kind='line', data=exercise)
```

```python
sns.catplot(x='time', y='pulse', kind='violin', data=exercise)
```

```python
sns.catplot(x='time', y='pulse', kind='box', data=exercise);
```

```python
sns.relplot(x='time', y='pulse', hue='kind', style='diet', data=exercise)
```
