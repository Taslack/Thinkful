```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Loading one of the example datasets from seaborn.
exercise = sns.load_dataset('exercise')
```

```python
lowfatrest1min = exercise.loc[(exercise['diet'] == 'low fat') & (exercise['kind'] == 'rest') & (exercise['time'] == '1 min'), 'pulse']
lowfatrest15min = exercise.loc[(exercise['diet'] == 'low fat') & (exercise['kind'] == 'rest') & (exercise['time'] == '15 min'), 'pulse']
lowfatrest30min = exercise.loc[(exercise['diet'] == 'low fat') & (exercise['kind'] == 'rest') & (exercise['time'] == '30 min'), 'pulse']
```

```python
sns.relplot(x='time', y='pulse', data=exercise)
```

Pros:</br>
- Good separation pulse and time.</br>
- Can see good indication of what pulse does as time increases.</br>

Cons:</br>
- No separation of what type of diet they are on, can not tell what effect that has on pulse and time.</br>
- Grouping of data points can be clustered.</br>

```python
sns.relplot(x='time', y='pulse', kind='line', data=exercise)
```

Pros:</br>
- Good indication of the mean.</br>
- Shows the 95% confidence interval.</br>

Cons:</br>
- Does not fit this data, provides less information.</br>
- Can not determine if there is any correlation between time and pulse.


```python
sns.catplot(x='time', y='pulse', kind='violin', data=exercise)
```

Pros:</br>
- See the distribution of pulse values separated by min.</br>
- Indication of the mean is represented.</br>
- Can see the how heart rate changes over time.</br>
- Good representation of mean and median.</br>

Cons:</br>
- Can not see specific data points.</br>

```python
sns.catplot(x='time', y='pulse', kind='box', data=exercise);
```

Pros:</br>
- Gives good representation of min, max, and median.</br>
- Displays outlier data points.</br>
- Can see how pulse changes over the minuets.</br>
- Displays upper and lower quartiles. </br>

Cons:</br>
- Exact values not retained.</br>

```python
sns.relplot(x='time', y='pulse', hue='kind', style='diet', data=exercise)
```

Pros:</br>
- See the distribution of pulse values separated by min.</br>
- Indication of the mean is represented.</br>
- Can see the how heart rate changes over time.</br>
- Really see the difference to pulse in teh no fat and low fat diets.</br>
- Separates out the type of exercise and how it effects pulse vs min.

Cons:</br>
- Data points can be clustered and unreadable.</br>
- Not really great separation of the no fat and low fat values, hard to tell which is which.

```python
iris = sns.load_dataset('iris')
sns.relplot(x='sepal_length', y='petal_width', data=iris)
```

Pros:</br>
- You can tell there is a correlation with petal_width and sepal_length</br>
- Grouping of data points can be seen easily.</br>
- Good for seeing outliers.</br>

Cons:</br>
- The stacking of data points can occur, causing overplotting.</br>

```python
plt.figure(figsize=(16, 6))
sns.barplot(x='sepal_length', y='petal_width', data=iris)
```

Pros:</br>
- Groups all datapoints together, give an error between the difference in datapoints compared to the y axis.</br>

Cons:</br>
- Can not see any relationship between datapoints or the x and y axis.</br>

```python
plt.figure(figsize=(16, 6))
sns.boxplot(x='sepal_length', y='petal_width', data=iris)
```

Pros:</br>
- Shows outliers.</br>
- Gives you a good indication of min, max, and median.</br>
- Can see that as sepal_length increases so does the petal_width.</br>

Cons:</br>
- Individual datapoints are not visible.</br>

```python
attention = sns.load_dataset('attention')
sns.relplot(x='attention', y='score', data=attention)
```

Pros:</br>
- Good separation of data.</br>
- Able to see the differences attention has on score.</br>

Cons:</br>
- Can cluster datapoints to be unrecognizable.</br>
- No indication of min, max, or median.</br>
- Does not indicate the amount of datapoints in each score.

```python
sns.boxplot(x='attention', y='score', data=attention)
```

Pros:</br>
- Can see min, max, and median clearly.</br>
- Shows how higher score is correlated with focused attention.</br>

Cons:</br>
- No display of actual datapoints.</br>

```python
sns.barplot(x='attention', y='score', data=attention)
```

Pros:</br>
- Stacks datapoints and gives an error.</br>

Cons:</br>
- Lower end of data is obscured by bar.</br>
- Error indication does not provide much information.</br>

```python
sns.violinplot(x='attention', y='score', data=attention)
```

Pros:</br>
- Gives good max, min, and median indications.</br>
- Can see how the data is distributed.</br>
- Indicates the difference between types of attentions.</br>
- Can see how score is effected by the types of attention.</br>

Cons:</br>
- Can not see exact data.</br>
- Mean indicator is small.</br>

```python
sns.catplot(x='attention', y='score', data=attention)
```

Pros:</br>
- Can see the distribution of all the datapoints.</br>
- Gives a clear indication of score vs attention.</br>

Cons:</br>
- Nothing that indicates the max, min, or mean.</br>

```python
sns.swarmplot(x='attention', y='score', data=attention)
```

Pros:</br>
- Organizes all the datapoints for clear visibility.</br>
- Can see the effect of score vs attention</br>

Cons:
- No max, min, or mean.</br>
