```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('global-and-regional-population-estimates-us-census-bureau-vs-un.csv', index_col='Year', usecols=['Year', 'Entity', 'United Nations ( )'])
df.rename(columns={'United Nations ( )': 'Population'}, inplace=True)
df = df.pivot(columns='Entity')
df['Population']['United States'].plot()
df['Population']['Africa'].plot()
df['Population']['Europe'].plot()
df['Population']['South America'].plot()

ax = df['Population']['Asia'].plot()
ax.set_ylabel("Population (In Millions)")
ax.legend()
ax.plot()
```

```python
ax = df['Population']['China'].plot()
ax.axvspan(1979, 2015, facecolor='r', alpha=0.5)
ax.set_ylabel("Population (In Millions)")
ax.legend()
```

This is Chinas population since 1950, the red is when Chinas one child policy was being enforced.  You can see a decline in population growth happening after about 10 years from the start of the program.

```python
ax = df['Population']['India'].plot()
ax = df['Population']['China'].plot()
ax.axvspan(1979, 2015, facecolor='g', alpha=0.5)
ax.set_ylabel("Population (In Millions)")
ax.legend()
```

Here is India and Chinas popluation for the same period marked in green.  Indias population maintained a constant growth rate and did not decrease like Chinas. 

```python
sy = df.loc[2015]
labels = 'United States', 'Africa', 'Europe', 'South America', 'Asia'
pops = [sy['Population']['United States'], sy['Population']['Africa'], sy['Population']['Europe'], sy['Population']['South America'], sy['Population']['Asia']]

fig1, ax1 = plt.subplots()
ax1.pie(pops, labels=labels, autopct='%1.1f%%', pctdistance=.8)
plt.title('2015 Population per Continent')
plt.show()
```

A good representation on what continent has the most populace.
