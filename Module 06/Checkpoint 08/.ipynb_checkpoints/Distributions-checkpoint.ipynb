```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
r = np.random.pareto(10, 100)
plt.hist(r, color='g')
plt.axvline(r.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(r.mean() + r.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(r.mean() - r.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

Most of the data falls within the standard deviation.  And the mean is a good representation of where the center is.

```python
r = np.random.rayleigh(1, 100)
plt.hist(r, color='g')
plt.axvline(r.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(r.mean() + r.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(r.mean() - r.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

It has a good represention of how the data is spread out.

```python
r = np.random.hypergeometric(1, 1, 1, 100)
plt.hist(r, color='g')
plt.axvline(r.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(r.mean() + r.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(r.mean() - r.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

This is skewed, most of the data does not fall within 1 of the standard deviation.

```python
r = np.random.exponential(1, 100)
plt.hist(r, color='g')
plt.axvline(r.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(r.mean() + r.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(r.mean() - r.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

Most of the data falls within 1 of the standard deviation, but the data is skewed.

```python
r = np.random.triangular(0, 1, 2, 100)
plt.hist(r, color='g')
plt.axvline(r.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(r.mean() + r.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(r.mean() - r.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

The mean is a good indicator of where most of the data is falling and most of the data is falling within the standard deviation.

```python
r = np.random.f(10, 10, 100)
plt.hist(r, color='g')
plt.axvline(r.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(r.mean() + r.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(r.mean() - r.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

The grouping is mostly towards zero, this data is skewed.

```python
s = np.random.normal(5, .5, 100)
l = np.random.normal(10, 1, 100)
r = s + l
plt.hist(r, color='g')
plt.axvline(r.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(r.mean() + r.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(r.mean() - r.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

This data is a good set, falling with in the center of the mean and 1 of the standard deviation.
